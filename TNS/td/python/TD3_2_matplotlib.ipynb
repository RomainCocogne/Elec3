## 3. Tracer une sinusoïde avec des axes centrés

Tracez la même sinusoïde avec les axes centrés sur (0,0), cette fois-ci démarrant en $t$ = -15 ms et allant jusque $t$ =15 ms et avec la légende "$sin(2\pi.f.t+\phi)$"
.
Pour les titres des axes, voir https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.text pour les placer "à la main".


```python
# -*- coding: utf-8 -*-
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(-0.015, 0.015, 1/(5E6));
y = 2.5*np.sin(100*x*2*np.pi+50*np.pi/180)
plt.plot(x, y,label="sin(2*pi*f*t+phi)")
plt.title('un sinus')
plt.legend(["sin(2pift+phi)"],loc='upper right')
ax = plt.gca() 
plt.xlabel("amplitude")
ax.xaxis.set_label_coords(0.5,-0.05)
plt.ylabel("temps")
ax.yaxis.set_label_coords(0,0.5)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

# 5.  Créer une fonction

Créez une fonction "plotsin" avec comme paramètres :
   amplitude
   fréquence
   phase           (en degrés)
   temps de début
   temps de fin
   fréquence d'échantillonnage
   discret/continu
   titre
   
   
Testez la avec différents paramètres, en particulier avec une fréquence de 100 Hz et une fréquence d'échantillonnage de 200 Hz.  Cela devrait vous amener à utiliser xlim et ylim.
   

```python
# -*- coding: utf-8 -*-
import numpy as np
from matplotlib import pyplot as plt

def plotsin(ampl,f,phi,t0,tf,freq,c,titre):
    x = np.arange(t0, tf, 1/(freq));
    y = ampl*np.sin(f*x*2*np.pi+phi*np.pi/180)
    if c!="continue":
        plt.stem(x, y,label="sin(2*pi*f*t+phi)")
    if c!="discret":
        plt.plot(x, y,label="sin(2*pi*f*t+phi)")
    plt.title(titre)
    plt.legend(["sin(2pift+phi)"],loc='upper right')
    ax = plt.gca() 
    plt.xlabel("amplitude")
    ax.xaxis.set_label_coords(0.5,-0.05)
    plt.ylabel("temps")
    ax.yaxis.set_label_coords(0,0.5)
    
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))

plt.subplot(211)
plotsin(2.5,100,0,-0.015,0.015,500,"continue","f=500hz")
plt.subplot(212)
plotsin(0.5,100,0,-0.015,0.015,500,"discret","f=500hz")
```
