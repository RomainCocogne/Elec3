```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import scipy
import scipy.fftpack as fftpack

%matplotlib inline 

img = cv2.imread("lena.png",0) 
```

```python
def plot (img,title,limite=1):
    if(np.size(img.shape)==2):
        plt.figure(1,figsize=(10,10))
        plt.subplot(111), plt.imshow(np.abs(img), cmap = 'gray',vmax=np.max(img)*limite,vmin=0)
        plt.title(title + str(img.shape)) 
        plt.show()
    else:
        plt.figure(1,figsize=(10,10))
        plt.subplot(111), plt.plot(img)
        plt.show()
        
def spectre(img):
    if(np.size(img.shape)==2):
        f = np.fft.fft2(img)
        fshift = np.fft.fftshift(f)
    else:
        f = np.fft.fft(img)
        fshift = np.fft.fftshift(f)
    return fshift
```

 1) Pour réduire significativement la taille de l’image, quelles sont à votre avis les informations d’une image qui pourraient être éliminées sans trop dégrader le contenu de l’image.

On choisit de supprimer les composantes haute fréquence de l'image, car elles sont de toute façon imperceptibles à l'oeil nu.

On utilise la DCT pour isoler les hautes fréquences, car nous n'avons pas besoin de la phase et la dct est plus rapide que la fft (qui fait des calculs complexes).

```python
def dct(img):
    F=fftpack.dct(fftpack.dct(img,norm="ortho",axis=0),norm="ortho",axis=1)
    return F
```

```python
plot(dct(img),"dct de lena.png",limite=1e-3)
plot(np.abs(spectre(img)),"fft de lena.png",limite=1e-3)
```

#### On voit bien ici que la DCT et la FFT donnent les mêmes résultats mais que la FFT calcule le spectre en entier alors que ce dernier est périodique. Dans ce sens, la DCT est plus optimisée.

3)Est-il préférable de travailler avec des blocs de pixels (de taille 8x8 par exemple) ou
bien sur une image entière ? Appliquer l’outil mathématique identifié à la question précédente sur des blocs de pixels 8x8.

#### Il est préférable de travailler avec des blocs de 8x8 plutôt que sur l'image entière. En effet, on va supprimer les hautes fréquences dans un environnement restreint où l'image est quasi-stationnaire, afin d'éviter, par exemple, que la fréquence du bord droit de l'image n'affecte le bord gauche. 

#### De plus, cela réduit les calculs, car le calcul des coefficients ne se fera plus qu'avec 8x8 pixels.

```python
def dct_parse(img):
    while (img.shape[0]%8 !=0):
        img=np.concatenate((img, img[-1,:].reshape(1,img.shape[1])),axis=0)
        
    while (img.shape[1]%8 !=0):
        img=np.concatenate((img, img[:,-1].reshape(img.shape[0],1)),axis=1)
    print(img.shape)
    
    F=np.zeros(img.shape);
    for octet_x in np.arange(0,img.shape[0],8):
        for octet_y in np.arange(0,img.shape[1],8): 
            square=img[octet_x:octet_x+8,octet_y:octet_y+8]
            F[octet_x:octet_x+8,octet_y:octet_y+8]=dct(square)
    return F
```

```python
plot(dct_parse(img),"dct 8x8 lena.png",limite=1e-3)
```
