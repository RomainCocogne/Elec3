```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img_base = cv2.imread("D:\Polytech\Projet_2\Docs\lena.png",0) #on importe l'image test

plt.figure(1,figsize=(10,10)) #on cre une figure de taille fixe (10 par 10)
plt.subplot(111), plt.imshow(img_base, cmap = 'gray')
plt.title("Image de base " + str(img_base.shape)) #on affiche le nombre de pixel de l'image

plt.show()
```

```python
#On fait la fft double (spectre 2D) de l'image de base
fft_base = np.fft.fft2(img_base)
#On centre le spectre
fft_shift_base = np.fft.fftshift(fft_base)
    
[lignes_base, colonnes_base] = img_base.shape #on rcupre le nombre de pixel en x et en y
    
X = np.arange(-lignes_base/2, lignes_base/2,1)/lignes_base #Normalisation
Y = np.arange(-colonnes_base/2, colonnes_base/2,1)/colonnes_base #Normalisation
[x,y] = np.meshgrid(X,Y)
    
L=10
alpha=np.pi

flou = np.sinc(np.pi*L*(x*np.cos(alpha)+y*np.sin(alpha)));


flou_dB = 20*np.log(np.abs(flou)+ 0.001) #module


#on affiche le spectre du filtre "flou" 
plt.figure(1,figsize=(10,10)), plt.imshow(flou_dB, cmap='gray')
plt.title("Filtre de bouge"), plt.xticks([]), plt.yticks([])

#on affiche l'image avec le filtre flou apliqu
spectre_img_floutee = flou * fft_shift_base
ifft_shift_img_floutee = flou * fft_shift_base #filtrer correspond a faire la multiplication du filtre par la FFTshift de l'image en frquentiel
img_floutee = np.fft.ifft2(ifft_shift_img_floutee) #Transforme 2d inverse pour rcuperer notre image floute dans le domaine spatial

plt.figure(2,figsize=(10,10)), plt.imshow(np.abs(img_floutee), cmap='gray')
plt.title("Image floutee")


#On fait la fft double (spectre 2D) de l'image de base
fft_2 = np.fft.fft2(img_floutee)
#On centre le spectre
fft_shift_2 = np.fft.fftshift(fft_2)


#on affiche l'image deflout
filtre = 1/flou #on fait ici le filtre pour rectifier notre flou de boug
spectre_img_filtree = filtre * fft_shift_2 #on fait la multiplication en frquentiel
ifft_shift_img_filtree = np.fft.ifftshift(spectre_img_filtree)
img_filtree = np.fft.ifft2(ifft_shift_img_filtree) #enfin on retourne en spatial
plt.figure(3,figsize=(10,10)), plt.imshow(np.abs(img_filtree), cmap='gray')
plt.title('Image filtree')





plt.show()
```
